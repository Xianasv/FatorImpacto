```python
import pandas as pd
from itertools import chain
from ast import literal_eval
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pickle

def write_data_to_disk(file, data):
    with open(file, 'wb') as fid:
        pickle.dump(data, fid)


def load_data_from_disk(file):
    with open(file, 'rb') as fid:
        data = pickle.load(fid)
    return data
```

**Função que gera a lista de pacs que têm 50 artigos como mínimo**

```python
def dic_pacs_50art_years(revista):
    dw = dz[dz.journal == revista] #art correct
    dv = ds[ds.journal == revista] #art repetidos busca pacs

    years = list(dw.year.unique())
    years.sort()

    Pacs_nro_50art_x = {}   # nro pacs c/>= 50 art c/ escolha no ano x
    Pacs_list_50art_x = {}  # list pacs c/>= 50 art c/ escolha no ano x
    for year in years:
        print(" ")
        print('x: ', year)

        dy = dv[dv.year == year]  # criterio de escolha - aqui fijamos os pacs a buscar por nro de art
        dy1 = dv[dv.year == year - 1] 
        dy2 = dv[dv.year == year - 2]

        print(year - 1)
    #     print('dy1', dy1['pacs-n3-u'])
        print(year - 2 )
    #     print('dy2', dy2['pacs-n3-u'])
        print('------------------------------------------')

        gp = list(dy['pacs-n3-u'].unique())
        gp.sort()

        pac_50art_x = []
        for pac in gp:
            # print(pac)
            dy1_art = dy1[dy1['pacs-n3-u'] == pac]
            dy2_art = dy2[dy2['pacs-n3-u'] == pac]
            ny1_art = len(dy1_art.UT)
            ny2_art = len(dy2_art.UT)

            nro_art_x2 = ny1_art + ny2_art  #nro art do pac nos anos x-1, x-2 
            # print(nro_art_x2)

            if nro_art_x2 >= 50:
                # print('pac: ', pac)
                # print(nro_art_x2)
                pac_50art_x.append(pac)

        nro_pac50art_x = len(pac_50art_x)
        print(pac_50art_x, nro_pac50art_x)
        print('------------------------------------------')

        if nro_pac50art_x != 0:
            Pacs_nro_50art_x[year-2, year-1] = nro_pac50art_x
            Pacs_list_50art_x[year] = pac_50art_x

    print(Pacs_nro_50art_x) # guardar info ** para 30 e 20 também Urg!
    print('-.-.-.-.-.-.-.-.-.-.-.-.-.-.-.-.-.-.-.-.-.-.-.-.-.-.-.')
    print(Pacs_list_50art_x)
    print('**********************************************')
    
    return Pacs_list_50art_x
    
```
